```python
import requests
from lxml import html
from langdetect import detect, detect_langs

url = "https://en.wikisource.org/wiki/Grammar_of_the_Burmese_Language"

# Get content of the website
response = requests.get(url)
web_content = response.content
# Parse the HTML content
tree = html.fromstring(web_content)

# Extract text only
text_content = tree.xpath('//body//text()')
# Join the list of strings into a single string
full_text = ' '.join(text_content)

# Detect the language of the content
language = detect(full_text)
languages_detected = detect_langs(full_text)

print(f"Detected Language: {language}")
print(f"Languages Detected: {languages_detected}")
```
